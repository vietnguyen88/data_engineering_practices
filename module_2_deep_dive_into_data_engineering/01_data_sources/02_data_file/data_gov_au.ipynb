Data File - Excel

```python
import pandas as pd
import os
# input variables
excel_file_path = "auditor_and_liquidator.xlsx"
auditor_sheetname = "REG_AUDITOR_202511"
liquidator_sheetname = "LIQUIDATOR_202510"

# output variables
path = "data_lake"
```

Read Auditor data

```python
# You code here
# 1. Read data from "REG_AUDITOR_202511" spreadsheet using Pandas
df = pd.read_excel(excel_file_path,sheet_name=auditor_sheetname,header=2)
# 2. Apply data type conversion and export to CSV
# print(f"""
#       ------------------
#       Before convert
#       ------------------  
# {df.dtypes}""")

converted_df = df.convert_dtypes(infer_objects=True)
# print(f"""
#       ------------------
#       After convert
#       ------------------ 
# {converted_df.dtypes}""")
# # 3. Export to path 'destination/excel'
path = 'destination/excel'
os.makedirs(path,exist_ok=True)
print('Exporting CSV...')
converted_df.to_csv(f'{path}/converted_dtypes.csv')
csv_exist = os.path.exists(f'{path}/converted_dtypes.csv')
if csv_exist:
    print('Finish')
    
```

Read Liquidator data

```python
# You code here
# 1. Read data from "LIQUIDATOR_202510" spreadsheet using Pandas
#    + Provide a schema when reading data file
liquidator_schema = {
    "REGISTER_NAME": str,
    "LIQ_NUM": str,
    "OFF_LIQ_NUM": str, 
    "LIQ_NAME": str,
    "LIQ_START_DT": 'datetime64[ns]',
    "OFF_LIQ_START_DT": 'datetime64[ns]',
    "LIQ_STATUS": str,
    "LIQ_SUSP_DT": 'datetime64[ns]',
    "LIQ_ADD_LOCAL": str,
    "LIQ_ADD_STATE": str,
    "LIQ_ADD_PCODE": str,
    "LIQ_ADD_COUNTRY": str,
    "LIQ_FIRM": str
}
df = pd.read_excel(excel_file_path,sheet_name=liquidator_sheetname,dtype=liquidator_schema)
df.head(5)
# df.dtypes
# 2. Export to CSV as path 'destination/excel'
df.to_csv(f'{path}/schema_dtypes.csv')
```
