1. Simple Structure

```python
import pandas as pd
import os
import json
```

```python

# HINTS
path = 'destination/json'
simple_json_file_path = "simple_structure.json"
# 1. Use pandas.read_json()
df = pd.read_json(simple_json_file_path)
os.makedirs(path,exist_ok=True)
# 2. Export to CSV as path: 'destination/json'
df.to_csv(f'{path}/simple_structure.csv')
```

2. Multiple Level

```python

# HINTS
# 1. Use pandas json_normalize to read data from record_path
# 2. Export to CSV as path: 'destination/json'

multiple_level_json_file_path = "multiple_level.json"

f = open(multiple_level_json_file_path,'r')
data = json.load(f)
# print(data)
df = pd.json_normalize(data,record_path='students')
df.to_csv(f'{path}/multiple_level.csv')
```

3. Nested Levels

```python
# HINTS
nested_level_json_file_path = "nested_level.json"
# Approach 1: use record_path + meta
# 1. Open the load JSON into dictionary 
f = open(nested_level_json_file_path,'r')
data = json.load(f)
# print(data)
# 2. Use pandas json_normalize to read data from record_path
df = pd.json_normalize(data,record_path='students',meta=['school_name',['class','level'],['class','room'],['class','teacher']])
df.head()
# 3. Export to CSV as path: 'destination/json'
# df.to_csv(f'{path}/nested_level.csv')
```
